### Question 6
The term "day trade" refers to a financial market trading strategy in which an investor buys and sells assets on the same day. The main objective of day trading is to capitalize on short-term fluctuations in asset prices, seeking to profit from intraday variations.

But when carrying out such a strategy, the Brazilian stock exchange (B3) modifies the trading fees that are applied to each of the transactions based on the volume (in R$ million) that a single person trades. Since after each R$ 1 Mi that is traded, a discount of `n * 0.05%` is applied to the trading fee, which starts at 0.005% of the volume invested, with n being the value of the volume. Example:

|volume(R$ million)|Negotiation rate|
|------------------|---------------|
|0|0.0050%|
|1|0.00475%|
|2|0.0045%|
|3|0.00425%|
|4|0.004%|
|5|0.00375%|

Write a program in Python that receives from the user the trading volume in millions (1: R$1,000,000.00, 2: R$2,000,000.00). Your program must return the rate that was applied and the value, in reais, of the fee that will be paid by the user based on the traded volume entered.

Attention: The user must only enter integer values, and it is necessary to carry out the appropriate checks, asking for the entry again, when necessary

Example:

```python
Entrance:
Trading volume: 1
---
Exit:
Rate applied: 0.00475%
Fee value: R$ 4750.00
```

````python
negotiation = int(input("Enter the value of the negotiation, in millions (0 to 5): "))

if 0 <= negotiation <= 5:
    base_rate = 0.005
    discount_per_million = 0.0005

    applied_rate = base_rate - (negotiation * discount_per_million)

    rate_value = negotiation * applied_rate

    print(f"\nApplied rate: {applied_rate * 100:.5f}%")
    print(f"Fee value: R$ {rate_value:.2f}")
else:
    print("Please enter a value between 0 and 5.")
````

### Question 7
When dealing with stock exchange trading data, it is important to keep a record of all transactions that each investor carries out.

You need to start thinking about implementing a solution that stores the different trade values that are provided by the person using your Python program. You must store, in a list, the values as they are provided by the user.

With each request for a new trade value, you must ask the user if she wants to enter a new trade, with the answer being 'Y' for "Yes", and 'N' for "No". Thus, when the user types "N", their program must be terminated (that is, stop receiving new values).

After building the list (that is, after receiving all the values that the user wants to enter), print on the screen:

All values entered by the user, in the order in which they were entered
All values entered by the user, in ascending order
The average value of trades
Only values greater than the average
Only values smaller than the average
The number of values entered

````python
negotiations = []

while True:
    value = float(input("Enter the trade value: "))
    negotiations.append(value)

    continues = input("Insert a new negotiation? (Y for Yes, N for No): ")
    if continues.upper() == 'N':
        break

print("Values entered by the user (in order of insertion):")
print(negotiations)

for value in negotiations:
     print(f"Value: {value}")

average = sum(negotiations) / len(negotiations)
print(f"Average value of trades: {average}")

largest_avg = [valor for valor in negotiations if valor > average]
print("Values greater than average:")
print(largest_avg)

smallest_avg = [valor for valor in negotiations if valor < average]
print("Values less than average:")
print(smallest_avg)

quantity = len(negotiations)
print(f"Number of values entered: {quantity}")
````
