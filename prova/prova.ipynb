### Questão 6
O termo "day trade" refere-se a uma estratégia de negociação no mercado financeiro em que um investidor compra e vende ativos no mesmo dia. O objetivo principal do day trade é capitalizar as flutuações de curto prazo nos preços dos ativos, buscando lucrar com as variações intradiárias.

Mas ao realizar tal estratégia, a bolsa de valores do Brasil (B3) modifica as taxas de negociações que são aplicadas para cada uma das transações com base no volume (em R$ milhões) que uma única pessoa negocia. Sendo que após cada R$ 1 Mi que é negociado, é aplicado um desconto de `n * 0,05%` na taxa de negociação, que se inicia como sendo 0.005% do volume investido, sendo n o valor do volume. Exemplo:

|volume(R$ milhões)|Taxa negociação|
|------------------|---------------|
|0|0.0050%|
|1|0.00475%|
|2|0.0045%|
|3|0.00425%|
|4|0.004%|
|5|0.00375%|

Escreva um programa em python que receba da pessoa usuária, o volume de negociação em milhões (1: R$ 1.000.000,00, 2: R $ 2.000.000,00). Seu programa deve retornar a taxa que foi aplicada e o valor, em reais, de taxa que será paga pelo usuário com base no volume negociado inserido.

Atenção: O usuário deverá digitar apenas valores inteiros, sendo necessário realizar as verificações apropriadas, pedindo novamente pela entrada, quando necessário

Exemplo:

```python
Entrada:
Volume da negociação: 1
---
Saida:
Taxa aplicada: 0.00475%
Valor da taxa: R$ 4750.00
```

````python
negociacao = int(input("Digite o valor da negociação, em milhões (0 a 5): "))

if 0 <= negociacao <= 5:
    taxa_base = 0.005
    desconto_por_milhao = 0.0005

    taxa_aplicada = taxa_base - (negociacao * desconto_por_milhao)

    valor_taxa = negociacao * taxa_aplicada

    print(f"\nTaxa aplicada: {taxa_aplicada * 100:.5f}%")
    print(f"Valor da taxa: R$ {valor_taxa:.2f}")
else:
    print("Por favor, insira um valor entre 0 e 5.")
````

### Questão 7
Ao lidarmos com dados de negociações da bolsa de valores, é importante mantermos o registro de todas as transações que cada um dos investidores realiza.

Você precisa começar a pensar na implementação de uma solução que armazene os diferentes valores de negociações que são fornecidos pela pessoa usuária do seu programa em Python. Você deve armazenar, em uma lista, os valores conforme eles forem sendo fornecidos pela pessoa usuária.

A cada solicitação de um novo valor de negociação, você deve perguntar à usuária se ela deseja inserir uma nova negociação, tendo, como resposta, 'S' para "Sim", e 'N', para "Não". Assim, quando a pessoa usuária digita "N", seu programa deve ser encerrado (isto é, parar de receber novos valores).

Após construir a lista (ou seja, após receber todos os valores que a pessoa usuária deseja inserir), imprima na tela:

Todos os valores digitados pela usuária, na ordem em que foram inseridos
Todos os valores digitados pela usuária, em ordem crescente
O valor médio das negociações
Apenas os valores maiores que a média
Apenas os valores menores que a média
A quantidade de valores digitados

````python
negociacoes = []

while True:
    valor = float(input("Digite o valor da negociação: "))
    negociacoes.append(valor)

    continua = input("Inserir uma nova negociação? (S para Sim, N para Não): ")
    if continua.upper() == 'N':
        break

print("Valores digitados pela usuária (na ordem de inserção):")
print(negociacoes)

for valor in negociacoes:
    print(f"Valor: {valor}")

media = sum(negociacoes) / len(negociacoes)
print(f"Valor médio das negociações: {media}")

maiores_media = [valor for valor in negociacoes if valor > media]
print("Valores maiores que a média:")
print(maiores_media)

menores_media = [valor for valor in negociacoes if valor < media]
print("Valores menores que a média:")
print(menores_media)

quantidade_valores = len(negociacoes)
print(f"Quantidade de valores digitados: {quantidade_valores}")
````

### Questão 8
Uma nota de negociação na bolsa de valores se refera a um documento que detalha as informações essenciais de uma transação financeira realizada no mercado. Essas notas são registros formais que documentam as operações de compra e venda de ativos, oferecendo uma visão clara e organizada das transações realizadas por um investidor ou uma instituição financeira.

Considere que você precisa implementar um protótipo de uma solução, em Python, que, a partir da entrada de determinados valores, imprima métricas características de uma nota de negociação. O seu programa deve inicialmente ter como entrada:

Código da empresa (string com o código de negociação na bolsa de valores de uma empresa)
Quantidade de ações compradas (inteiro)
Preço da ação no momento que a compra foi realizada (float que indica o preço de uma única ação)
Já como saída, seu programa deve:

Lista com o nome de todas as ações compras
Preço médio das ações presente nessa nota de negociação
Valor total da nota de negociação (valor calculado com base no preço de cada ação e quantidade de ações compradas = `∑(quantidade_acoes∗preco_medio_acao))`

Exemplo:

```python
Entrada:

1 iteração de entrada:
Código da Empresa: B3SA3
Quantidade de ações compradas: 100
Preço da ação no momento da compra: 12.66

2 iteração de entrada:
Código da empresa: VALE3
Quantidade de ações compradas: 200
Preço da ação no momento da compra: 73.61

---
Saída
Lista de todas empresas compradas: ['B3SA3', 'VALE3']
Preço médio das ações presentes nessa nota de negociação: R$ 43.135
Valor total da nota de negociação: R$ 15988.0
```


**Atenção: No caso de quantidade de ações compradas, a usuária deverá digitar apenas valores inteiros, sendo necessário realizar as verificações apropriadas, pedindo novamente pela entrada, quando necessário**

````python
empresas_compradas = []
quantidade_acoes_total = 0
valor_total_nota = 0.0

while True:
    codigo_empresa = input('Código: ')
    while True:
        try:
            quantidade_acoes = int(input('Quantidade de ações compradas: '))
            if quantidade_acoes <= 0:
                print('A quantidade de ações deve ser um número inteiro positivo.')
                continue
            break
        except ValueError:
            print('A quantidade de ações deve ser um número inteiro.')

    while True:
        try:
            preco_acao = float(input('Preço da ação no momento da compra: '))
            if preco_acao <= 0:
                print('O preço da ação deve ser um número positivo.')
                continue
            break
        except ValueError:
            print('O preço da ação deve ser um número.')

    empresas_compradas.append(codigo_empresa)
    quantidade_acoes_total += quantidade_acoes
    valor_total_nota += quantidade_acoes * preco_acao

    resposta = input('Deseja adicionar mais uma empresa à nota de negociação? (S/N): ')
    if resposta.lower() != 's':
        break

if quantidade_acoes_total > 0:
    preco_medio = valor_total_nota / quantidade_acoes_total
    print('\nLista de todas empresas compradas:', empresas_compradas)
    print('Preço médio das ações presentes nessa nota de negociação: R$ {:.3f}'.format(preco_medio))
    print('Valor total da nota de negociação: R$ {:.2f}'.format(valor_total_nota))
else:
    print('Nenhuma ação foi adicionada à nota de negociação.')
````
